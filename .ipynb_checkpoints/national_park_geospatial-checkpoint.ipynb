```python
import folium
```

```python
# Create map object

m = folium.Map(location=[39.50, -98.35], zoom_start=10)
```

```python
# Global tooltip

tooltip = 'Click For More'
```

```python
# Create custome marker icon

logoIcon = folium.features.CustomIcon('nrg_logo.png')
```

```python
# Create Markers

# New River Gorge National Park - West Virginia

folium.Marker([37.8707, -81.0531],
              popup='<strong>New River Gorge National Park</strong>',
              tooltip=tooltip,).add_to(m)
```

```python
folium.Marker([44.3386, -68.2733],
              popup='<strong>Acadia National Park</strong>',
              tooltip=tooltip).add_to(m)
```
